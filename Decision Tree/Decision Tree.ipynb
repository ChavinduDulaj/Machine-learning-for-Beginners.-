```python
import pandas as pd
df = pd.read_csv('salaries.csv')
df.head()
```

# Preprocession the data using encoding

```python
inputs = df.drop('salary_more_then_100k', axis='columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n=inputs.drop(['company','job','degree'], axis='columns')
```

```python
inputs_n
```

# Train dataset using Decision tree calssifier

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```

# Predic the person who works in google as sales  executive with master degree

```python
model.predict([[2,2,1]])
```

# Execise for predicting the survival rate of passenger in titanic ship

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
inputs = df.drop('Survived',axis='columns')
inputs
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2}) #mapping the sex as 0 or 1 based on male or female
inputs
```

```python
target = df['Survived']
target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3,1,22,7.66]])
```
