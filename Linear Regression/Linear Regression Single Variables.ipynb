```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("dataset\homeprices.csv")
df
```

# Ploting the data using pyplot

```python

# Scatter plot
plt.scatter(df.area, df.price, color='red', marker='+')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Area vs Price')

# Display the plot
plt.show()
```

# Training the dataset using Linear Regression model

```python


# Assuming reg is your trained linear regression model
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df['price'])

# Reshape the input to a 2D array
input_value = np.array(3300).reshape(1, -1)

# Make the prediction
prediction = reg.predict(input_value)

# Print the prediction
print(prediction)
```

# ploting again after the prediction

```python

# Scatter plot
plt.scatter(df.area, df.price, color='red', marker='+')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Area vs Price')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
# Display the plot
plt.show()
```

# Checking the coefficients and intercept values

```python
reg.coef_
```

```python
reg.intercept_
```

```python
m = reg.coef_[0] #In scikit-learn, the coef_ attribute returns an array of coefficients when there are multiple features. In your case, you have a single feature ('area'), so the coef_ attribute returns an array with a single element.
c = reg.intercept_

# Get input from the user
x = float(input('Enter area for a land: '))

# Calculate the predicted price
y = (m * x) + c

print(f'Price for {x} area is {y}')
```

```python
# Reshape the input to a 2D array
input_value = np.array(5000).reshape(1, -1)

# Make the prediction
prediction = reg.predict(input_value)

# Print the prediction
print(prediction)
```

```python
df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)
```

```python
df.head(10)
```

```python

# Scatter plot
plt.scatter(df.year, df.income, color='red')
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('Scatter Plot of Year vs Income')

# Display the plot
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['income'])

user_input = input("Enter a year: ")

# Convert the user input to a numeric value
try:
    input_value = np.array(float(user_input)).reshape(1, -1)
except ValueError:
    print("Invalid input. Please enter a numeric value.")
    

# Make the prediction
prediction = reg.predict(input_value)

print(f"Predicted income for the year {user_input}: {prediction[0]}")


print(prediction)
```

```python

# Scatter plot
plt.scatter(df.year, df.income, color='red', marker='+')
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('Scatter Plot of Year vs Income')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')  # Use 'year' instead of 'income'
# Display the plot
plt.show()
```
