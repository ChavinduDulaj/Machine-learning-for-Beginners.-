```python
df[df.price_per_sqft<min_thresold]
```

```python
df2 = df[(df.price_per_sqft>min_thresold) & (df.price_per_sqft < max_thresold)]
```

```python
df2[df2.price_per_sqft<min_thresold]
```

```python
df2.shape
```

```python
df2.head(10)
```

```python
price_prer_quare_check=df2.price_per_sqft.unique()
has_string_value = any(isinstance(value, str) for value in price_prer_quare_check)

if has_string_value:
    print("String value(s) found in 'price_per_sqft' column:")
    for value in price_prer_quare_check:
        if isinstance(value, str):
            print(value)
else:
    print("No string values found in 'price_per_sqft' column.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Create a box plot for the DataFrame
sns.boxplot(data=df2.price_per_sqft)

# Show the plot
plt.show()

# Identify potential outliers for removal
q1 = df2['price_per_sqft'].quantile(0.25)
q3 = df2['price_per_sqft'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df2[(df2['price_per_sqft'] < lower_bound) | (df2['price_per_sqft'] > upper_bound)]

# Highlight the outliers on the plot
plt.scatter(outliers.index, outliers['price_per_sqft'], color='red', label='Outliers')

# Show the plot
plt.legend()
plt.show()
```
